# Air flight in USA in January 2020

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
flights_2019 = pd.read_csv('Jan_2019_ontime.csv')
flights_2020 = pd.read_csv('Jan_2020_ontime.csv')
```

```python
flights_2020 = flights_2020.loc[:, ~flights_2020.columns.str.contains('^Unnamed')]
flights_2019 = flights_2019.loc[:, ~flights_2019.columns.str.contains('^Unnamed')]
```

```python
flights_2020.head()
```

```python
flights_2020.columns
```

Unique values of airports

```python
airports = flights_2020.ORIGIN.unique()
```

```python
flights_2020.dropna(inplace=True)
flights_2019.dropna(inplace=True)
```

```python
flights_2020.DEP_DEL15.unique()
```

```python
flights_2020['DEPART_DELAYED'] = np.where(flights_2020.ARR_DEL15 > 0.0, 'delayed', 'on time')
flights_2019['DEPART_DELAYED'] = np.where(flights_2019.ARR_DEL15 > 0.0, 'delayed', 'on time')
flights_2020['ARRIVAL_DELAYED'] = np.where(flights_2020.DEP_DEL15 > 0.0, 'delayed', 'on time')
flights_2019['ARRIVAL_DELAYED'] = np.where(flights_2019.DEP_DEL15 > 0.0, 'delayed', 'on time')
```

```python
flights_2020.hist(figsize=(20, 15))
plt.suptitle('Histograms of raw data')
plt.show()
```

## Least Delayed Airports:

Those are the best 50 airports in having less departure delay 

```python
performances = {}
for airport in airports:
    flights_per_airport = flights_2020[flights_2020.ORIGIN == airport]
    delayed = flights_per_airport[flights_per_airport.DEPART_DELAYED == 'delayed'].shape[0]
    # divide the number of delayed trips over the total number per airport
    airport_performance = delayed / len(flights_per_airport)
    performances[airport] = airport_performance
    # sorting the values of the performances dictionary
    performances = {k: v for k, v in sorted(performances.items(), key=lambda item: item[1])}
```

Convert the dictionary to dataframe

```python
performances = pd.DataFrame.from_dict(performances, orient='index')
```

```python
plt.figure(figsize=(15,8))
plt.bar(performances[:50].index, performances[:50].iloc[:,0], color='green')
plt.xticks(rotation='vertical')
plt.xlabel('Least Delayed Airport')
plt.ylabel('Departure Delay Percentage')
plt.show()
```

The best airport in the list is UIN (Quincy Regional Airport) and COD (Yellowstone Regional Airport). It was not possible to get the full list name of airport names since there is 351 unique Origin airports unless it's done manually, So we kept using the IATA code for airport. 

## Most delayed airports

```python
plt.figure(figsize=(15,8))
plt.bar(performances[-50:].index, performances[-50:].iloc[:,0], color='green')
plt.xticks(rotation='vertical')
plt.xlabel('Most Delayed Airport')
plt.ylabel('Departure Delay Percentage')
plt.show()
```

Airport SHR (Sheridan County Airport) has almost 50% of the flights are delayed

## Count Departures delay per Time block

```python
time_df = flights_2020[flights_2020.DEP_DEL15 != 0.0]
time_df = time_df[['DEP_TIME_BLK', 'DEP_DEL15']]
time_df = time_df.sort_values('DEP_TIME_BLK',ascending=True)
time_df = time_df[['DEP_TIME_BLK', 'DEP_DEL15']].groupby(['DEP_TIME_BLK']).agg(['count'])
```

```python
time_df.columns
```

```python
plt.figure(figsize=(15,8))
x = time_df.index
y = time_df.iloc[:,0]
plt.bar(x, y, color='green')
plt.xticks(rotation='vertical')
plt.xlabel('Time Block')
plt.ylabel('Frequency of Delays')
plt.show()
```
